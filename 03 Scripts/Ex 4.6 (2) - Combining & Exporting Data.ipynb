```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

### 1. Import the data sets into Jupyter

```python
# Turn project folder path into a string.
path = '/Users/asadagha/Documents/CF Data Analytics/Data Analytics Immersion/Achievement 4/Instacart Basket Analysis'
```

```python
# Import dataset orders_products_combined.pkl
df_ords_prods_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Import dataset products_checked.csv
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

### 2. Check the dimensions of the imported dataframes

```python
# Check the structure of the dataframe
df_ords_prods_combined.head()
```

```python
# Drop unnecessary columns
df_prods = df_prods.drop(['Unnamed: 0'], axis=1)
```

```python
# Check the structure of the dataframe
df_prods.head()
```

```python
df_prods.shape
```

### 3. Determine a suitable way to combine the orders_products_combined dataframe with the products data setÂ¶

```python
# Merge the 'orders_prods_combined' dataframe with 'df_prods'
df_ords_prods_merge = df_prods.merge(df_ords_prods_combined, on = 'product_id', indicator = 'merged_dataframes')
```

```python
# drop the flag column '_merge' from the previous merge
df_ords_prods_merge.drop('_merge', axis = 1 )
```

```python
df_ords_prods_merge.shape
```

### 4. Confirm the results of the merge using the merge flag

```python
df_ords_prods_merge['merged_dataframes'].value_counts()
```

### 5. Export the newly created dataframe as ords_prods_merge in a suitable format

```python
# Export data in pkl format
df_ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merge.pkl'))
```
