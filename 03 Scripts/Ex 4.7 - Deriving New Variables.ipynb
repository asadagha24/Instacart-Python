### Step 1a. Create "price_label" column.

```python
# Create a subset of first million rows

df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
df.head()
```

```python
# Define the 'price_labels' function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the 'price_labels' function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Find the maximum value in the 'prices' column

df['prices'].max()
```

```python
# Use the loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Use loc() function on the entire dataframe

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### Step 1b. Create "busiest_day" column.

```python
# Frequency of each value in the 'order_day_of_week' column

df_ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).

result = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in ords_prods_merge dataframe and set it equal to result

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### Step 2. Change the "Busiest day" column to "Busiest days" containing the two busiest days, and also create a new column for the "Slowest days" containing the two slowest days of the week. 

```python
# Create column “busiest_days" from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days).

result_2 = []

for value in df_ords_prods_merge["order_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

### 3. Check the values of this new column for accuracy.

```python
# Create new 'busiest_days' column in ords_prods_merge to view results in context.

df_ords_prods_merge['busiest_days'] = result_2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

### Step 4. Create new column "busiest_period_of_day" to categorize time periods labeled as “Most orders,” “Average orders,” and “Fewest orders.”

```python
# Check value counts in "order_hour_of_day" column.

df_ords_prods_merge['order_hour_of_day'].value_counts()
```

Splitting the value counts into equal thirds with the following labels: "Most orders" = 10, 11, 14, 15, 13, 12, 16, 9; "Fewest orders" = 23, 6, 0, 1, 5, 2, 4, 3; "Average orders" = all remaining values in order_hour_of_day.

```python
# Create for-loop labeling periods of time as “Most orders,” “Average orders,” and “Fewest orders.”

result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create a new column named "busiest_period_of_day" in the df_ords_prods_merge dataframe

df_ords_prods_merge['busiest_period_of_day'] = result_3
```

### Step 5. Print frequency of "busiest_period_of_day" column.

```python
# Print frequency count of values in the "busiest_period_of_day" column.

df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
