### Task 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Aggregating Data with transform() Function
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking the Output
ords_prods_merge.head(15)
```

```python
# Create 'Loyalty' Flags using loc Function
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check Value Counts of 'loyalty_flag'
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking the Updated Output
ords_prods_merge.head(15)
```

### Task 5. Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Determine basic statistics of customer spending habits based on 'loyalty_flag' categories
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

Those amongst the 'Loyal customer' category purchase items with a mean price of 10.39. Whereas those amongst the 'New customer' and 'Regular customer' categories both purchase items with a mean price of 13.29 and 12.50 respectively, both higher than that of items purchased by loyal customers.

### Task 6. Create a spending flag for each user based on the average price across all their orders using the following criteria:
1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender”
2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create a new column for the average spend based on the price of the products purchased by a customer
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check the new column
ords_prods_merge.head(15)
```

```python
# Assign either 'Low_spender' or 'High_spender' tag to each customer
ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'
ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check value counts of spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check Releavent Columns
ords_prods_merge[['user_id', 'average_spend', 'spender_flag']].head(15)
```

### Task 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a 'Non-frequent customer'
2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a 'Regular customer'
3. If the median is lower than or equal to 10, then the customer should be labeled a 'Frequent customer'

```python
# Create a new column for customer frequency flags based on the median values from 'days_since_prior_order' column
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Check the newly created column
ords_prods_merge.head(15)
```

```python
# Based on the criteria provided, create a column featuring order frequency flags
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['customer_frequency'] > 10) & (ords_prods_merge['customer_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts for new column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check columns of updated dataframe
ords_prods_merge[['user_id', 'customer_frequency', 'frequency_flag']].head(20)
```
