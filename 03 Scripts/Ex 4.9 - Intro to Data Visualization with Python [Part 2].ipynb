### Task 2. Create a bar chart, a histogram, a scatterplot, and a line chart, using a sample subset for the line chart.

#### 2(a) Bar Chart

```python
#Plot a bar chart
instacart['order_day_of_week'].value_counts().plot.bar()
```

```python
#Index the bar chart by order days of week
bar = instacart['order_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
#Add colors to the bar chart
bar = instacart['order_day_of_week'].value_counts().sort_index().plot.bar(color =['red','blue','purple','green','green','purple','yellow'])
```

```python
#Save the image
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_order_dow.png'))
```

#### 2(b) Histograms & Scatterplots

```python
#Create a histogram
instacart['prices'].plot.hist(bins = 25)
```

```python
#describing 'prices'
instacart['prices'].describe()
```

### Task 3. Provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
#Create a histogram
hist3 = instacart['order_hour_of_day'].plot.hist(bins = 24)
```

### Task 5. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day.

```python
#Create a sample of the dataframe
sampled_instacart = instacart.sample(n=100000, random_state=1)
```

```python
#Group by hour of the day and calculate the mean expenditure
hourly_expenditure = sampled_instacart.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
#Create a line chart
line_price_hod = sns.lineplot(data = sampled_instacart, x = 'order_hour_of_day', y = 'prices')
```

### Task 6. Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents.

```python
#Check columns
sampled_instacart.columns
```

```python
# Create a line chart
line_3 = sns.lineplot(data = sampled_instacart, x = 'age',y = 'n_dependants')
```

There is no distinct pattern that could indicate there being a connection between age and number of dependets.

### Task 7. Explore whether there’s a connection between age and spending power (income) using a scatterplot.

```python
#create a scatterplot for 'Age' and 'Income'
scatter = sns.scatterplot(x='age', y='income', data=sampled_instacart)
```

The scatterplot shows ages distribution ranging from 0 - 80 years old, and an income distribution from 0 - 600,000. The majority of customers' income is clustered in the range of 0 - 200,000. There is some variability in earnings that we can see but no clear trend can be observed from the scatterplot. We can see that some young customers (40 years or less) earn up to 400,000 while the variation in earnings amongst the older age group (40+ years) is higher with some customers earning up to 600,000.
